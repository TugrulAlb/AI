# Scaling - Normalization - PCA

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('pima-indians-diabetes.csv')
```

```python
df
```

```python
from sklearn.preprocessing import normalize, scale
```

```python
x = df.drop('Outcome', axis=1)
```

```python
x
```

```python
yeni_x = scale(x)
```

```python
pd.DataFrame(yeni_x)
```

```python
normalizedata = normalize(x)
```

```python
pd.DataFrame(normalizedata)
```

# Boyut Kucultme PCA - Principle Component Analysis

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(2)
```

```python
x.head()
```

```python
data2 = pca.fit_transform(x)
```

```python
data2
```

```python
pd.DataFrame(data2)
```

```python
pd.DataFrame(data2, columns = ['PCA1', 'PCA2'])
```
